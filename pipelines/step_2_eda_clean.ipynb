# Step 2: Processed CSV EDA and Cleaning Roadmap

This notebook loads the cleaned processed CSV outputs from `step_1_inspect.ipynb`, performs EDA, and documents the next cleaning/standardization steps.

```python
import pandas as pd
import os

processed_dir = os.path.join('..', 'data', 'processed')
source1_path = os.path.join('..', 'data', 'raw', 'source1_global.csv')
source2_path = os.path.join(processed_dir, 'source2_uk_clean.csv')
source3_path = os.path.join(processed_dir, 'source3_uk_clean.csv')

source1 = pd.read_csv(source1_path)
source2 = pd.read_csv(source2_path)
source3 = pd.read_csv(source3_path)

print('Loaded:', source1_path, '->', source1.shape)
print('Loaded:', source2_path, '->', source2.shape)
print('Loaded:', source3_path, '->', source3.shape)
```

## 2.1 Basic shape and preview

```python
print('Source 2 shape:', source2.shape)
print(source2.head(10).to_string(index=False))

print('\nSource 3 shape:', source3.shape)
print(source3.head(10).to_string(index=False))

print('\nSource 1 shape:', source1.shape)
print(source1.head(10).to_string(index=False))
```

## 2.2 Column and type summary

```python
print('Source 2 columns and dtypes:')
print(source2.dtypes)
print('\nSource 3 columns and dtypes:')
print(source3.dtypes)
print('\nSource 1 columns and dtypes:')
print(source1.dtypes)
```

## 2.3 Missing values and basic statistics

```python
print('Source 2 missing values:')
print(source2.isna().sum().sort_values(ascending=False).head(20))
print('\nSource 3 missing values:')
print(source3.isna().sum().sort_values(ascending=False).head(20))
print('\nSource 1 missing values:')
print(source1.isna().sum().sort_values(ascending=False).head(20))

print('\nSource 2 numeric summary:')
print(source2.describe(include='all').transpose())
print('\nSource 3 numeric summary:')
print(source3.describe(include='all').transpose())
print('\nSource 1 numeric summary:')
print(source1.describe(include='all').transpose())
```
