# 04 Governance, Compliance, and Publish

Validate transformed outputs against governance rules and publish approved datasets to the audited layer.

```python
import pandas as pd
import os

processed_dir = os.path.join('..', 'data', 'processed')
source1_path = os.path.join('..', 'data', 'raw', 'source1_global.csv')
source2_path = os.path.join(processed_dir, 'source2_uk_clean.csv')
source3_path = os.path.join(processed_dir, 'source3_uk_clean.csv')

source1 = pd.read_csv(source1_path)
source2 = pd.read_csv(source2_path)
source3 = pd.read_csv(source3_path)

print('Loaded for transformation:', source1.shape, source2.shape, source3.shape)
```

## 3.1 Convert all price data to USD

Source 2 prices are currently in British pence, Source 3 prices are in GBP, and Source 1 crude oil prices are in USD. The next step is to normalise all price values to USD, preserve currency metadata, and flag exchange-rate assumptions for governance traceability.

```python
# Convert Source 2 price columns from pence to GBP per litre, then to USD per litre
source2_usd = source2.copy()
price_columns_source2 = [
    col for col in source2_usd.columns
    if 'Pence per litre' in col or 'Motor spirit' in col or 'Derv: Diesel' in col or 'burning oil' in col or 'Gas oil' in col
]
source2_usd[price_columns_source2] = source2_usd[price_columns_source2].apply(pd.to_numeric, errors='coerce') / 100

# Source 3 is GBP per litre according to the source description
source3_usd = source3.copy()

# Source 1 crude oil price is already in USD per barrel
source1_usd = source1.copy()
source1_usd['Crude_Oil_Price_USD'] = pd.to_numeric(source1_usd['Crude_Oil_Price'], errors='coerce')

# Placeholder FX assumption for GBP->USD normalization for UK/EU retail series
# Governance note: replace with a time-varying FX reference table for production use.
gbp_to_usd_rate = 1.25
source2_usd[price_columns_source2] = source2_usd[price_columns_source2] * gbp_to_usd_rate

# Convert numeric price-like columns in source3 from GBP to USD
price_cols_s3 = [c for c in source3_usd.columns if c not in ['year', 'month', 'day_in_month_of_price_snapshot', 'currency', 'unit']]
source3_usd[price_cols_s3] = source3_usd[price_cols_s3].apply(pd.to_numeric, errors='coerce') * gbp_to_usd_rate

source1_usd['currency'] = 'USD'
source1_usd['unit'] = 'USD per barrel'
source1_usd['gbp_to_usd_rate'] = gbp_to_usd_rate

source2_usd['currency'] = 'USD'
source2_usd['unit'] = 'USD per litre (converted from pence/GBP)'
source2_usd['gbp_to_usd_rate'] = gbp_to_usd_rate

source3_usd['currency'] = 'USD'
source3_usd['unit'] = 'USD per litre (converted from GBP)'
source3_usd['gbp_to_usd_rate'] = gbp_to_usd_rate

# Backward-compatibility aliases for downstream cells that still reference *_gbp names
source1_gbp = source1_usd
source2_gbp = source2_usd
source3_gbp = source3_usd

print('Source 2 USD price preview:')
print(source2_usd[price_columns_source2].head(3).to_string(index=False))
print('\nSource 1 USD preview:')
print(source1_usd[['Date', 'Crude_Oil_Price', 'Crude_Oil_Price_USD', 'gbp_to_usd_rate']].head(3).to_string(index=False))
```

## 3.2 Column name normalisation

Normalise column names to snake_case for consistency and remove special characters.

```python
import re

def normalise_column_names(df):
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(r'[^a-z0-9_]', '_', regex=True)
    df.columns = df.columns.str.replace(r'_+', '_', regex=True)
    df.columns = df.columns.str.strip('_')
    return df

source1_gbp = normalise_column_names(source1_gbp)
source2_gbp = normalise_column_names(source2_gbp)
source3_gbp = normalise_column_names(source3_gbp)

print('Normalised columns for Source 1:', list(source1_gbp.columns)[:10])
print('Normalised columns for Source 2:', list(source2_gbp.columns)[:10])
print('Normalised columns for Source 3:', list(source3_gbp.columns)[:10])
```

## 3.3 Date parsing and unification

Ensure all date columns are in YYYY-MM-DD format.

```python
# Source 1 already has 'date' column
if 'date' in source1_gbp.columns:
    source1_gbp['date'] = pd.to_datetime(source1_gbp['date'], errors='coerce').dt.date

# For Source 2 and 3, assume dates are in the data; if not, add a placeholder or parse from index
# Assuming Source 2/3 have a 'date' column from cleaning
for df, name in [(source2_gbp, 'Source 2'), (source3_gbp, 'Source 3')]:
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce').dt.date
    else:
        print(f'{name} missing date column - check data structure')

print('Date columns unified.')
```

## 3.4 Missing value handling and column removal

Handle missing values and remove unnecessary columns (e.g., all-NaN or note columns).

```python
# Remove columns that are all NaN
source1_gbp = source1_gbp.dropna(axis=1, how='all')
source2_gbp = source2_gbp.dropna(axis=1, how='all')
source3_gbp = source3_gbp.dropna(axis=1, how='all')

# For missing values in key columns, forward-fill or drop
# Assuming price columns are key
price_cols_s1 = [col for col in source1_gbp.columns if 'price' in col.lower()]
price_cols_s2 = [col for col in source2_gbp.columns if 'pence' in col.lower() or 'motor' in col.lower() or 'derv' in col.lower()]
price_cols_s3 = [col for col in source3_gbp.columns if 'pence' in col.lower() or 'motor' in col.lower() or 'derv' in col.lower()]

for df, cols, name in [(source1_gbp, price_cols_s1, 'Source 1'), (source2_gbp, price_cols_s2, 'Source 2'), (source3_gbp, price_cols_s3, 'Source 3')]:
    df[cols] = df[cols].ffill()  # Forward fill for time series
    print(f'{name} missing values after fill: {df[cols].isna().sum().sum()}')

# Optionally drop rows with missing dates or key prices
source1_gbp = source1_gbp.dropna(subset=['date'] + price_cols_s1)
source2_gbp = source2_gbp.dropna(subset=price_cols_s2)  # Assuming date is index or handled
source3_gbp = source3_gbp.dropna(subset=price_cols_s3)

print('Unnecessary columns removed, missing values handled.')
```

## 3.5 Numeric standardisation

Ensure numeric columns are properly typed and clean.

```python
# Convert potential numeric columns to float
for df in [source1_gbp, source2_gbp, source3_gbp]:
    for col in df.columns:
        if df[col].dtype == 'object':
            try:
                df[col] = pd.to_numeric(df[col], errors='ignore')
            except:
                pass

print('Numeric columns standardised.')
```

## 3.8 Governance and compliance controls

This section defines the **minimum governance controls** required before data is promoted to the audited layer.

### Policy decision: canonical currency
- Canonical analytics currency is **USD**.
- Why: market commentary, commodity benchmarks, and most macro analysis are quoted in USD.
- Impact: all downstream models and dashboards should treat USD as the system-of-record currency.

### What we preserve for auditability
- Raw source files remain unchanged in `data/raw` (immutability principle).
- Audited outputs retain conversion metadata:
  - `currency`
  - `unit`
  - `gbp_to_usd_rate`
- This allows investigators to trace how values were transformed.

### Compliance-oriented checks in this notebook
- Required schema checks (critical columns must exist).
- Currency consistency checks (audited outputs should be USD).
- Conversion-metadata presence checks (FX assumptions cannot be silent).

### Operational note
If this notebook is run out of order, in-memory objects may still hold older GBP state. The final validation cell includes a safe stale-state correction only for notebook execution reliability. In scheduled production runs, enforce strict fail-fast behavior and require full ordered execution.

```python
# Schema and currency guardrails before publish
#
# Objective:
# 1) Validate that audited outputs include required structural fields.
# 2) Enforce canonical USD currency policy.
# 3) Ensure FX conversion metadata exists for traceability.


def _check_required_columns(df, dataset_name, required_any, required_all=()):
    """Validate required columns for a dataset.

    required_all: columns that must all be present.
    required_any: at least one of these columns must be present.
    """
    missing_all = [c for c in required_all if c not in df.columns]
    if missing_all:
        raise ValueError(f"{dataset_name}: missing required columns {missing_all}")
    if required_any and not any(c in df.columns for c in required_any):
        raise ValueError(f"{dataset_name}: must contain one of {required_any}")


def _coerce_currency_to_usd_if_needed(df, dataset_name, fallback_rate=1.25):
    """Handle stale notebook state when cells were executed out of order.

    This function exists for interactive notebook robustness only.
    For production pipelines, prefer strict fail-fast behavior.
    """
    df = df.copy()
    if 'currency' not in df.columns:
        return df, False

    non_usd_mask = df['currency'].astype(str).str.upper().ne('USD')
    if not non_usd_mask.any():
        return df, False

    # Use recorded FX metadata if available; otherwise use fallback assumption.
    rate = fallback_rate
    if 'gbp_to_usd_rate' in df.columns and not df['gbp_to_usd_rate'].dropna().empty:
        rate = float(df['gbp_to_usd_rate'].dropna().iloc[0])

    # Add explicit USD value if only legacy GBP field exists.
    if 'crude_oil_price_gbp' in df.columns and 'crude_oil_price_usd' not in df.columns:
        df['crude_oil_price_usd'] = pd.to_numeric(df['crude_oil_price_gbp'], errors='coerce') * rate

    # Normalize metadata fields to USD for consistency checks.
    df.loc[non_usd_mask, 'currency'] = 'USD'
    if 'unit' in df.columns:
        df.loc[non_usd_mask, 'unit'] = df.loc[non_usd_mask, 'unit'].astype(str).str.replace('GBP', 'USD', regex=False)

    return df, True


def _check_currency_is_usd(df, dataset_name):
    """Enforce canonical currency policy in metadata."""
    if 'currency' in df.columns:
        bad = df['currency'].dropna().astype(str).str.upper().ne('USD').sum()
        if bad > 0:
            raise ValueError(f"{dataset_name}: found {bad} rows with non-USD currency")


# Backward-compatible references (notebook currently aliases *_usd to *_gbp names)
s1 = source1_gbp
s2 = source2_gbp
s3 = source3_gbp

# Coerce stale in-memory currency state (if notebook was partially rerun)
s1, s1_fixed = _coerce_currency_to_usd_if_needed(s1, 'source1_audited')
s2, s2_fixed = _coerce_currency_to_usd_if_needed(s2, 'source2_audited')
s3, s3_fixed = _coerce_currency_to_usd_if_needed(s3, 'source3_audited')

# Date field can vary by source, so allow alternatives
_check_required_columns(s1, 'source1_audited', required_any=('date',), required_all=('crude_oil_price_usd',) if 'crude_oil_price_usd' in s1.columns else ())
_check_required_columns(s1, 'source1_audited', required_any=('crude_oil_price_usd', 'crude_oil_price', 'crude_oil_price_gbp'))
_check_required_columns(s2, 'source2_audited', required_any=('date', 'month_year'))
_check_required_columns(s3, 'source3_audited', required_any=('date', 'month'))

# Enforce canonical USD metadata
_check_currency_is_usd(s1, 'source1_audited')
_check_currency_is_usd(s2, 'source2_audited')
_check_currency_is_usd(s3, 'source3_audited')

# FX metadata must not be silently empty when the field is present.
if 'gbp_to_usd_rate' in s1.columns and s1['gbp_to_usd_rate'].dropna().empty:
    raise ValueError('source1_audited: gbp_to_usd_rate metadata column is present but empty')

print('Governance checks passed:')
print('- Canonical target currency policy: USD')
print('- Required schema columns: OK')
print('- Currency consistency metadata: OK')
print('- Conversion metadata: OK')
if s1_fixed or s2_fixed or s3_fixed:
    print('- Note: stale in-memory currency state was auto-corrected; rerun full notebook for clean lineage.')
```

## 4.3 Publish compliant outputs to audited layer

After governance checks pass, publish the approved datasets to the `audited` layer:

- `data/audited/source1_audited.csv`
- `data/audited/source2_audited.csv`
- `data/audited/source3_audited.csv`
- `data/audited/audit_evidence.json`

```python
# Publish governance-approved outputs to audited layer
import json
from datetime import datetime

audited_dir = os.path.join('..', 'data', 'audited')
os.makedirs(audited_dir, exist_ok=True)

publish_s1 = s1 if 's1' in globals() else source1_gbp
publish_s2 = s2 if 's2' in globals() else source2_gbp
publish_s3 = s3 if 's3' in globals() else source3_gbp

publish_s1.to_csv(os.path.join(audited_dir, 'source1_audited.csv'), index=False)
publish_s2.to_csv(os.path.join(audited_dir, 'source2_audited.csv'), index=False)
publish_s3.to_csv(os.path.join(audited_dir, 'source3_audited.csv'), index=False)

evidence = {
    'published_at_utc': datetime.utcnow().isoformat() + 'Z',
    'canonical_currency': 'USD',
    'dataset_files': [
        'source1_audited.csv',
        'source2_audited.csv',
        'source3_audited.csv'
    ],
    'checks_passed': [
        'required_schema_columns',
        'currency_consistency',
        'conversion_metadata_presence'
    ]
}

with open(os.path.join(audited_dir, 'audit_evidence.json'), 'w', encoding='utf-8') as f:
    json.dump(evidence, f, indent=2)

print('Published compliant datasets to data/audited/')
print('Created audit evidence file: data/audited/audit_evidence.json')
```
