# Step 3: Inspect Source 3 - Additional UK Fuel Data

```python
import pandas as pd

# Check available sheets in source3_uk.xlsx
xls = pd.ExcelFile("../data/raw/source3_uk.xlsx")
print("Sheets in source3_uk.xlsx:", xls.sheet_names)

# Load and inspect the main data sheet
df_raw3 = pd.read_excel("../data/raw/source3_uk.xlsx", sheet_name="5.1.1 (excl tax)", header=None)
print("Source 3 UK - Shape:", df_raw3.shape)
print("Sample data:")
df_raw3.head(20)
```

# Step 4: Convert Excel Files to CSV

```python
import os

# Run the conversion script
print("Converting Excel files to CSV...")
result = os.system("python convert_files.py")
if result == 0:
    print("Conversion completed successfully!")
else:
    print("Conversion failed!")

# Verify the converted files
processed_dir = "../data/processed"
if os.path.exists(processed_dir):
    csv_files = [f for f in os.listdir(processed_dir) if f.endswith('.csv')]
    print(f"Converted CSV files: {csv_files}")
else:
    print("Processed directory not found")
```
