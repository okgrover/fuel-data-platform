# Fuel Data Platform - Step 1: Inspect and Convert Raw Data Sources

This notebook performs step-by-step inspection of all raw data sources and converts Excel files to CSV format.

# Step 1: Inspect Source 1 - Global Crude Oil Prices

```python
import pandas as pd

# Load and inspect source1_global.csv
df_global = pd.read_csv("../data/raw/source1_global.csv")
print("Source 1 Global - Shape:", df_global.shape)
print("Columns:", df_global.columns.tolist())
print("Sample data:")
df_global.head()
```

# Step 2: Inspect Source 2 - UK Fuel Data

```python
import pandas as pd

# Load and inspect source2_uk.xlsx
df_raw2 = pd.read_excel("../data/raw/source2_uk.xlsx", sheet_name="4.1.1", header=None)
print("Source 2 UK - Shape:", df_raw2.shape)
print("Sample data:")
df_raw2.head(20)
```

# Step 3: Inspect Source 3 - Additional UK Fuel Data

```python
import pandas as pd

# Check available sheets in source3_uk.xlsx
xls = pd.ExcelFile("../data/raw/source3_uk.xlsx")
print("Sheets in source3_uk.xlsx:", xls.sheet_names)

# Load and inspect the main data sheet
df_raw3 = pd.read_excel("../data/raw/source3_uk.xlsx", sheet_name="5.1.1 (excl tax)", header=None)
print("Source 3 UK - Shape:", df_raw3.shape)
print("Sample data:")
df_raw3.head(20)
```

# Step 4: Trim to Relevant Rows and Save Clean CSVs

We have inspected the raw Excel files, and now we create cleaned CSV exports from the notebook itself:

- **Source 1**: already a CSV file, no trimming needed
- **Source 2**: use row 9 as header, row 10+ as data
- **Source 3**: use row 7 as header, row 8+ as data

This keeps `convert_files.py` simple and makes the row-selection logic explicit in the notebook.

```python
import os

processed_dir = "../data/processed"
os.makedirs(processed_dir, exist_ok=True)

# Source 2: extract header row and data rows after inspection
headers2 = df_raw2.iloc[9].fillna('').astype(str).tolist()
df2_clean = df_raw2.iloc[10:].reset_index(drop=True)
df2_clean.columns = headers2
df2_clean = df2_clean.dropna(how='all')
print('Source 2 cleaned shape:', df2_clean.shape)
print(df2_clean.head(5))

# Source 3: extract header row and data rows after inspection
headers3 = df_raw3.iloc[7].fillna('').astype(str).tolist()
df3_clean = df_raw3.iloc[8:].reset_index(drop=True)
df3_clean.columns = headers3
df3_clean = df3_clean.dropna(how='all')
print('Source 3 cleaned shape:', df3_clean.shape)
print(df3_clean.head(5))

# Save cleaned CSV files
source2_csv = os.path.join(processed_dir, 'source2_uk_clean.csv')
source3_csv = os.path.join(processed_dir, 'source3_uk_clean.csv')
df2_clean.to_csv(source2_csv, index=False)
df3_clean.to_csv(source3_csv, index=False)
print(f'Saved cleaned source2 CSV to: {source2_csv}')
print(f'Saved cleaned source3 CSV to: {source3_csv}')

# Verify output files
csv_files = [f for f in os.listdir(processed_dir) if f.endswith('.csv')]
print('Current processed CSV files:', csv_files)
```

# Step 5: Quick EDA on Processed CSV Files

Now that the CSV exports are created, we inspect the processed files to confirm what was captured.

```python
import pandas as pd

# Load cleaned processed files for quick EDA
processed_source2 = pd.read_csv(source2_csv)
processed_source3 = pd.read_csv(source3_csv)

print('Processed Source 2 shape:', processed_source2.shape)
print('Processed Source 2 columns:', processed_source2.columns.tolist())
print('Processed Source 2 head(10):')
print(processed_source2.head(10).to_string(index=False))

print('\nProcessed Source 3 shape:', processed_source3.shape)
print('Processed Source 3 columns:', processed_source3.columns.tolist())
print('Processed Source 3 head(10):')
print(processed_source3.head(10).to_string(index=False))
```
